```python
#MongoDB Assignment
"""Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
MongoDB over SQL databases?"""
Ans: MongoDB is a cross-platform document-oriented database program. Classified as a NoSQL database program, MongoDB uses 
JSON-like documents with schema. MongoDB is developed by MongoDB Inc. and is free and open-source, published under a 
combination of the GNU Affero General Public License and the Apache License.

Non-relational databases are a type of database that does not use the traditional table-based relational database structure. 
Instead, they store data in a more flexible and dynamic way, often using documents, key-value pairs, or graph structures. 
Non-relational databases are often used for applications that require scalability, flexibility, and high performance.

#MOngoDB VS SQL
1. When you need to store large amounts of unstructured data.
2. When you need to store data with dynamic schemas.
3. When you need to scale horizontally and have high availability requirements.
4. When you need to perform complex queries on your data.
5. When you need to store data in a document-oriented format.
6. When you need to quickly access data from multiple sources.
7. When you need to perform real-time analytics on your data.
8. When you need to store data in a distributed fashion.

"""Q2. State and Explain the features of MongoDB."""
Ans: MongoDB is a document-oriented NoSQL database that is used for storing and managing large datasets. It is an open-source 
database that provides high performance, scalability, and availability. MongoDB features include:

1. Document-oriented Storage: MongoDB stores data in documents, which are similar to JSON objects. Documents contain key-value 
pairs and can be nested, making it easy to store complex data structures.

2. Indexing: MongoDB supports indexing on any field in a document, allowing for faster query performance.

3. Replication: MongoDB supports replication, which allows for data to be stored across multiple servers for increased 
availability and scalability.

4. Sharding: MongoDB supports sharding, which allows for data to be distributed across multiple servers. This helps with 
scalability and performance.

5. Aggregation: MongoDB supports aggregation, which allows for data to be grouped and analyzed.

6. Security: MongoDB provides a number of security features, such as authentication, authorization, and encryption.

"""Q3. Write a code to connect MongoDB to Python. Also, create a database and a collection in MongoDB."""
Ans: If you are using jupyter notebook then you can install pymongo
#install pymongo
pip install pymongo

#importing the pymongo library
import pymongo

#creating a connection to MongoDB
client = pymongo.MongoClient("mongodb://localhost:2701/")

#creating a database in MongoDB 
db = myclient["pwskills"]

#creating a collection in MongoDB 
coll_pwskills = mydb["customers"]

"""Q4. Using the database and the collection created in question number 3, write a code to insert one record, and insert many 
records. Use the find() and find_one() methods to print the inserted record."""
Ans: 
    #importing the pymongo library
import pymongo

#creating a connection to MongoDB
client = pymongo.MongoClient("mongodb://localhost:2701/")

#creating a database in MongoDB 
db = myclient["pwskills"]

#data record
customer = {"name" : "amit singh" , 
        "class" :"data science masters " ,
        "time" : "flexi"}

#creating a collection in MongoDB 
coll_pwskills = db["customer"]

#insert one record in MongoDB
coll_pwskills.insert_one(customer)

#multiple record
customer = [{"name" : "amit singh" , 
        "class" :"data science masters " ,
        "time" : "flexi"},
        {"email_id" : "xyz@gmail.com" , 
        "id" :"8789 " ,
        "time" : "flexi"},
        {"name" : "akshay" , 
        "class" :"data science certification " ,
        "time" : "fixed"}]
#insert to many record in MongoDB
coll_pwskills.insert_many(customer)

#use find()
for i in coll_pwskills.find():
    print(i)
    
#use find one()
for i in coll_pwskills.find({'name': 'amit singh'}):
    print(i)
    
"""Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB."""
Ans: The sort() method in MongoDB is used to sort the documents in a collection based on a specified field or fields. 
The sort() method takes an argument which is a list of fields along with their respective sorting order. The sorting order can 
be either 1 for ascending order or -1 for descending order.

#For example:
db.users.find().sort({age: 1})

"""Q7. Explain why delete_one(), delete_many(), and drop() is used."""
Ans: 
(A) Delete_one() is used to delete a single document from a MongoDB collection. It takes a query parameter to specify which 
document to delete.

(B) Delete_many() is used to delete multiple documents from a MongoDB collection. It takes a query parameter to specify which 
documents to delete.

(C) Drop() is used to delete an entire collection from a MongoDB database. It takes no parameters and will delete all documents
in the collection.
```
